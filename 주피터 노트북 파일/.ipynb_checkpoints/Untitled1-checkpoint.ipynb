```python
df = pd.read_csv('charge.csv')
df.head()
```

```python
df = df.filter(['statnm', 'lat', 'lng'])
df.head()
```

```python
# zoom_start: 배율 1~22
map_osm = folium.Map(location=[37.566651, 126.978428], zoom_start=12)
map_osm
```

```python
geo_path = 'geo.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
for i in df.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = df.loc[i, 'statnm']
    lat = df.loc[i, 'lat']
    lng = df.loc[i, 'lng']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name)
    marker.add_to(map_osm)
    
map_osm.choropleth(geo_data = geo_str)
    
map_osm.save('sra.html')
```
